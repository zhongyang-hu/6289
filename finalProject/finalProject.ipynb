```python
data1=pd.read_csv('https://raw.githubusercontent.com/zhongyang-hu/6289/main/finalProject/model1.csv')
data2=pd.read_csv('https://raw.githubusercontent.com/zhongyang-hu/6289/main/finalProject/model2.csv')
data3=pd.read_csv('https://raw.githubusercontent.com/zhongyang-hu/6289/main/finalProject/model3.csv')
data4=pd.read_csv('https://raw.githubusercontent.com/zhongyang-hu/6289/main/finalProject/model4.csv')
```

```python
x=[i for i in range(100)]
plt.plot(x,data1['val_sparse_categorical_accuracy'],label='LSTM model 1')
plt.plot(x,data2['val_sparse_categorical_accuracy'],label='LSTM model 2')
plt.plot(x,data3['val_sparse_categorical_accuracy'],label='LSTM model 1 with balanced data')
plt.plot(x,data4['val_accuracy'],label='Transformer')
plt.title('Test accuracy')

plt.legend()
```

```python
x=[i for i in range(100)]
plt.plot(x,data1['sparse_categorical_accuracy'],label='LSTM model 1')
plt.plot(x,data2['sparse_categorical_accuracy'],label='LSTM model 2')
plt.plot(x,data3['sparse_categorical_accuracy'],label='LSTM model 1 with balanced data')
plt.plot(x,data4['accuracy'],label='Transformer')
plt.title('Train accuracy')


plt.legend()
```
