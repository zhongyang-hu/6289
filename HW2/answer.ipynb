In coding part, I use two methods to use sigmoid function in CNN. First method, I use sigmoid function in full connected layer. It shows the lower accuracy than ReLu CNN.
Second method, I use sigmoid also in convoluted layer. In this part the accuracy is extremely low in first 10 epochs

Since the sigmoid function's outputs have the range from 0 to 1. In convoluted layer, the filters are used to figure out the features from images. And use sigmoid function will make the features in convoluted layers not significant since it limited the range to express the difference of features.In method one, the negative impact from applying the sigmoid function in full connected layer still exists. For the output, we want the result to be in range of 0 to 1.
In hidden layer, sigmoid may still shrink the range to express the information of difference.
